```python
from datasets import load_dataset

# Load the dataset
prompt_data = load_dataset("nomic-ai/gpt4all-j-prompt-generations", split='train')

# Filter the dataset: retain only entries where the 'prompt' length is 20 characters or fewer
filtered_data = prompt_data.filter(lambda example: len(example['prompt']) <= 25)

# Save the filtered dataset to a CSV file
filtered_data.to_csv("filtered_prompt_data.csv")

# Print the number of rows saved to confirm
print("Saved {} rows where 'prompt' length is 30 characters or fewer.".format(filtered_data.num_rows))                          
```

```python
#join data and azureData on prompt_length and ContextTokens
data['prompt_length'] = data['prompt_length'].astype(int)
azureData['ContextTokens'] = azureData['ContextTokens'].astype(int)

merged = data.merge(azureData, left_on='prompt_length', right_on='ContextTokens', how='inner')


```

```python
#write merged to a csv file
merged.to_csv("merged_data.csv")
```

```python
#only cosider rows with unique prompts
unique_prompts = merged.drop_duplicates(subset='prompt')
unique_prompts.head()
```

```python
#add a column called llm indicating the llm used to generate the prompt
# it should be one of the following: gpt-2, gemma-7b, llama-2-7b-chat-hf, mistral-7b-instruct
# randoml distribute the llm used to generate the prompt for each row
import random
llm = ['gpt-2', 'gemma-7b', 'llama-2-7b-chat-hf', 'mistral-7b-instruct']
unique_prompts['llm'] = random.choices(llm, k=unique_prompts.shape[0])

unique_prompts.head()
```

```python
#reset index
unique_prompts.reset_index(drop=True, inplace=True)
unique_prompts.head()
```

```python
from ray.data import from_pandas
import pandas as pd

local_model_paths = {
    "gpt-2": "/home/peiman/projects/RayInference/gpt-2",
    "llama-2-7b-chat-hf": "/home/peiman/projects/RayInference/llama-2-7b-chat-hf",
    "gemma-7b": "/home/peiman/projects/RayInference/gemma-7b",
    "mistral-7b-instruct": "/home/peiman/projects/RayInference/mistral-7b-instruct",
}
print(local_model_paths.items())

data_path = '/home/peiman/projects/RayInference/test12.csv'
data_df = pd.read_csv(data_path)
data = from_pandas(data_df)

# Map model names onto data
data = data.map_batches(
    lambda df: df.assign(model_name=df['llm'].apply(lambda x: local_model_paths.get(x, None))),
    batch_size=5,  # Adjust batch size as needed
    batch_format='pandas'
)

for model_name, path in local_model_paths.items():
    model_data = data.filter(lambda x: x['model_name'] == path)
    print(model_data.count())
    #save the model data to a csv file
    model_data.to_pandas().to_csv(f"{model_name}.csv")
```
